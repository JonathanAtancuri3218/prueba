```python
from tkinter import *
from tkinter import messagebox
import sqlite3

root= Tk()
root.geometry("300x450")

#===Funciones===#
def conexionBase():
	conexion= sqlite3.connect("BaseDatos3")
	cursor= conexion.cursor()

	try:
		cursor.execute('''
			CREATE TABLE CLIENTES(
			ID INTEGER PRIMARY KEY, 
			NOMBRE VARCHAR(10), 
			DIRECCION INTEGER(10), 
			TELEFONO VARCHAR(10), 
			EMAIL VARCHAR(20))
			''')

		messagebox.showinfo("Base de Datos3", "La base de datos se ha creado correctamente")

	except:
		messagebox.showwarning("Warning", "La base de datos ya esta creada")

def salirAplicacion():
	valor= messagebox.askquestion("Salir", "Seguro que deseas salir de la aplicacion?")

	if valor=="yes":
		root.destroy()

def limpiarCampos():
	id.set("")
	nombre.set("")
	direccion.set("")
	telefono.set("")
	email.set("")


def create():
	conexion= sqlite3.connect("BaseDatos3")
	cursor= conexion.cursor()

	"""cursor.execute("INSERT INTO CLIENTES VALUES('"+ id.get() + 
		"','" + nombre.get() + 
		"','" + direccion.get() +
		"','" + telefono.get() + 
		"','" + email.get() + "')")"""

	dates= id.get(), nombre.get(), direccion.get(), telefono.get(), email.get()

	cursor.execute("INSERT INTO CLIENTES VALUES(?,?,?,?,?)", dates)

	conexion.commit()

	messagebox.showinfo("Base de Datos", "Registro insertado con exito")

def read():
	conexion= sqlite3.connect("BaseDatos3")
	cursor= conexion.cursor()

	cursor.execute("SELECT * FROM CLIENTES WHERE ID=" + id.get())

	usuario= cursor.fetchall()

	for user in usuario:
		id.set(user[0])
		nombre.set(user[1])
		direccion.set(user[2])
		telefono.set(user[3])
		email.set(user[4])

	conexion.commit()


def update():
	conexion= sqlite3.connect("BaseDatos3")
	cursor= conexion.cursor()

	"""cursor.execute("UPDATE CLIENTES SET NOMBRE='" + nombre.get() +
		"', DIRECCION= '"+ direccion.get() +
		"', TELEFONO= '"+ telefono.get() +
		"', EMAIL= '"+ email.get() +
		"' WHERE ID=" + id.get()
		)"""

	dates= nombre.get(), direccion.get(), telefono.get(), email.get()

	cursor.execute("UPDATE CLIENTES SET NOMBRE=?, DIRECCION=?, TELEFONO=?, EMAIL=?" + 
					"WHERE ID=" + id.get(), dates)

	conexion.commit()

	messagebox.showinfo("Update", "Registro actualizado con exito")


def delete():
	conexion= sqlite3.connect("BaseDatos3")
	cursor= conexion.cursor()

	cursor.execute("DELETE FROM CLIENTES WHERE ID=" + id.get())

	conexion.commit()

	messagebox.showinfo("Delete", "Registro eliminado con exito")


#===Menu===#
barraMenu= Menu(root)
root.config(menu=barraMenu, width=300, height=300)

bbddMenu= Menu(barraMenu, tearoff= 0)
bbddMenu.add_command(label= "Crear BBDD", command=conexionBase)
bbddMenu.add_command(label= "Salir", command=salirAplicacion)

borrarMenu= Menu(barraMenu, tearoff= 0)
borrarMenu.add_command(label= "Borrar campos", command=limpiarCampos)

crudMenu= Menu(barraMenu, tearoff= 0)
crudMenu.add_command(label= "Crear", command= create)
crudMenu.add_command(label= "Buscar", command= read)
crudMenu.add_command(label= "Actualizar", command= update)
crudMenu.add_command(label= "Borar", command= delete)


barraMenu.add_cascade(label= "BBDD", menu=bbddMenu)
barraMenu.add_cascade(label= "Borrar", menu=borrarMenu)
barraMenu.add_cascade(label= "CRUD", menu=crudMenu)


#===Comienzo===#

frame= Frame(root)
frame.pack()

id= StringVar()
nombre= StringVar()
apellido= StringVar()
telefono= StringVar()
direccion= StringVar()
email= StringVar()


cuadroID= Entry(frame, textvariable= id) #textvariable-> Asi podremos modificar los string de Entry()
cuadroID.grid(row=0, column=1, padx=10, pady=10)

cuadroNombre= Entry(frame, textvariable= nombre)
cuadroNombre.grid(row=1, column=1, padx=10, pady=10)

cuadroDireccion= Entry(frame, textvariable= direccion)
cuadroDireccion.grid(row=2, column=1, padx=10, pady=10)

cuadroTelefono= Entry(frame, textvariable= telefono)
cuadroTelefono.grid(row=3, column=1, padx=10, pady=10)

cuadroEmail= Entry(frame, textvariable= email)
cuadroEmail.grid(row=4, column=1, padx=10, pady=10)


#===Etiquetas===#

idLabel= Label(frame, text= "ID : ")
idLabel.grid(row=0, column=0)

nombreLabel= Label(frame, text="Nombre : ")
nombreLabel.grid(row=1, column=0)

direccionLabel= Label(frame, text="Direccion : ")
direccionLabel.grid(row=2, column=0)

telefonoLabel= Label(frame, text="Telefono : ")
telefonoLabel.grid(row=3, column=0)

emailLabel= Label(frame, text="Email : ")
emailLabel.grid(row=4, column=0)



#===Botones===#

frame2= Frame(root)
frame2.pack()

botonCreate= Button(frame2, text="Crear", command=create)
botonCreate.grid(row=0, column=0, padx=10, pady=10)

botonRead= Button(frame2, text="Buscar", command= read)
botonRead.grid(row=0, column=1, padx=10, pady=10)

botonUpdate= Button(frame2, text="Actualizar", command= update)
botonUpdate.grid(row=0, column=2, padx=10, pady=10)

botonDelete= Button(frame2, text="Borrar", command= delete)
botonDelete.grid(row=0, column=3, padx=10, pady=10)




root.mainloop()
```
